# **Теоретическая часть**

## **Доказать, что оптимальные направления PCA совпадают с собственными векторами матрицы ковариаций.**

### **Определения:**
  * **PCA (Principal Component Analysis)** — это метод снижения
 размерности данных, основанный на поиске новых ортогональных осей, которые максимизируют дисперсию проекции
 данных. Эти новые оси называются главными компонентами.

 Первая компонента объясняет максимальную дисперсию данных. Каждая последующая компонента объясняет максимальную оставшуюся дисперсию при условии ортогональности предыдущим.

  * **Ковариация** между двумя векторами **x** и **у** (признаками) определяется как: $$cov(x,y)=\frac{1}{n} \sum_{i=1}^n (x_i- \mu_x)(y_i- \mu_y),$$ где $\mu_x$ и $\mu_y$ - средние значения.

  * **Дисперсия** - это ковариация признака с самим собой: $$var(x)=cov(x,x)=\frac{1}{n} \sum_{i=1}^n (x_i- \mu_x)^2$$

  * **Ковариационная матрица $\sum$** для данных **Х** (размер $n*p$): $$\sum = \frac{1}{n}X^TX,$$ где **Х** центрирована (столбцы имеют нулевое среднее).

  Элемент $\sum_{ij}$ - ковариация между $i$-м и $j$-м признаками.

  * **Формула $v^T \sum v$:** для единичного вектора $v(||v|| = 1)$: $$v^T \sum v$$ - это дисперсия проекции данных на направление **v**.

  Пояснение: Проекция данных **Х** на **v**: **$z=Xv$**. Тогда: $$var(z)=\frac{1}{n}z^Tz=\frac{1}{n}v^TX^TXv=v^T \sum v.$$

### **Доказательство:**
 * **Определение:** Ковариационная матрица набора данных фиксирует дисперсию и ковариацию между различными признаками данных. Для датасета с $n$ наблюдениями и $p$ признаками ковариационная матрица представляет собой матрицу $\sum$ размером $p × p$, где каждый элемент $\sum_{ij}$ соответствует ковариации между признаком $i$ и признаком $j$.

 * **Собственные значения и собственные векторы:**
 1. Для квадратной матрицы $\textbf{A}$ собственный вектор $v$ и его соответствующее собственное значение $\lambda$ удовлетворяют уравнению: $$Av= \lambda v$$

 В контексте ковариационной матрицы собственные векторы представляют направления в пространстве признаков, а собственные значения указывают величину дисперсии вдоль этих направлений.

 * **Анализ главных компонент (РСА):**

 1. **Преобразование в главные компоненты:** При выполнении РСА цель состоит в том, чтобы уменьшить размерность данных, сохранив при этом как можно больше дисперсию. Это достигается путем проецирования данных на направления (собственные векторы), которые захватывают наибольшую дисперсию.

 2. **Максимизация дисперсии:** Первая главная компонента - это собственный вектор, соответствующий наибольшему собственному значению ковариационной матрицы. Этот собственный вектор указывает направление, в котором данные изменяются больше всего. Последующие главные компоненты - это собственные векторы, соответствующие следующим по величине собственным значениям, гарантируя, что каждая новая компонента захватывает максимальную оставшуюся дисперсию, ортогональную предыдущим компонентам.

  **Пояснение:** Для любого единичного собственного вектора $v$ ковариационной матрицы $\sum$ верно: $v^T \sum v = \lambda$, где $\lambda$ - соответствующее собственное значение.

  По определению собственного вектора: $\sum v = \lambda v$. Умножаем слева на $v^T$: $v^T \sum v = v^T (\lambda v)=\lambda(v^Tv) = \lambda \cdot 1 = \lambda$ (так как $||v|| = 1$).

  Следовательно, максимизация $v^T \sum v$ сводится к выбору собственного вектора с наибольшим собственным значением.
  

 * **Почему собственные векторы являются главными компонентами:**

 1. **Геометрическая интерпретация:** Каждый собственный вектор можно рассматривать как новую ось в преобразованной системе координат. Проецируя исходные данные на эти оси, мы эффективно переориентируем данные так, чтобы они выравнивались по направлениям максимальной дисперсии. Это преобразование упрощает структуру данных, снижает размерность, сохраняя при этом ключевые характеристики.

 2. **Ортогональность:** Собственные векторы ковариационной матрицы ортогональны друг другу, что означает отсутствие корреляции между главными компонентами. Это свойство крайне важно для независимой интерпретации каждой компоненты.

  **Доказательство:** $\sum$ - ковариационная матрица, симметричная. $v_1, v_2$ - два ее различных собственных вектора, соответствующие разным собственным значением $\lambda_1, \lambda_2$.

  Рассмотрим скалярное произведение $v_1^T \sum v_2$ двумя способами:

  **1ый способ:** Так как $v_2$ собственный вектор: $v_1^T \sum v_2 = v_1^T (\lambda_2 v_2) = \lambda_2 (v_1^T v_2)$.

  **2ой способ:** Используем симметричность $\sum : (\sum^T=\sum)$, и то, что $v_1$ - собственный вектор: $v_1^T \sum v_2 = (\sum v_1)^T v_2 = (\lambda_1 v_1)^T v_2=\lambda_1(v_1^Tv_2).$

  **Получаем:** $\lambda_2 (v_1^T v_2) = \lambda_1(v_1^Tv_2)$

  **Переносим в одну сторону:** $(\lambda_2 - \lambda_1)(v_1^Tv_2)=0$

  По условию $\lambda_1 \neq \lambda_2$, значит: $v_1^Tv_2=0$.

 * **Вывод:** Собственные векторы ковариационной матрицы являются основными компонентами набора данных, поскольку они представляют направления максимальной дисперсии в данных. Проецируя данные на эти собственные векторы, можно эффективно уменьшить размерность, сохраняя при этом наиболее значимые особенности исходного набора данных. Поэтому РСА, который использует собственные векторы и собственные значения ковариационной матрицы, является мощным инструментом для анализа данных и уменьшения размерности.

```python
from IPython.display import display, Math, Latex
import math
from typing import List, Tuple, Optional, Any
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
from random import gauss
import random
```

```python
class Matrix:
    def __init__(self, n, m, matrix):
        self.n, self.m = n, m
        self.data, self.ind, self.indptr = self.to_csr(matrix)

    def to_csr(self, matrix):
        data = []
        ind = []
        indptr = [1]

        for i in range(self.n):
            count_nonzero = 0
            for j in range(self.m):
                value = matrix[i][j]
                if value != 0:
                    data.append(value)
                    ind.append(j)
                    count_nonzero += 1
            indptr.append(indptr[-1] + count_nonzero)

        return data, ind, indptr

    def from_csr(self):
        rows = self.n
        cols = self.m
        matrix = [[0 for _ in range(cols)] for _ in range(rows)]

        for i in range(rows):
            for j in range(self.indptr[i] - 1, self.indptr[i + 1] - 1):
                matrix[i][self.ind[j]] = self.data[j]

        return matrix

    def trace(self):
        if self.n != self.m:
            raise ValueError('Матрица не квадратная')

        trace_sum = 0
        for i in range(self.n):
            if 0 <= i < self.n:
                trace_sum += self.get_element(i + 1, i + 1)

        return trace_sum

    def get_element(self, row, col):
        row -= 1
        col -= 1
        if row < 0 or row >= self.n or col < 0 or col >= self.m:
            raise ValueError('Индексы выходят за пределы размеров матрицы')

        start = self.indptr[row] - 1
        end = self.indptr[row + 1] - 1
        for j in range(start, end):
            if self.ind[j] == col:
                return self.data[j]

        return 0

    def add(self, other):
        if self.n != other.n or self.m != other.m:
            raise ValueError('Размеры введенных матриц не совпадают')

        result_matrix = [[0] * self.m for _ in range(self.n)]

        for i in range(self.n):
            self_row_start = self.indptr[i] - 1
            self_row_end = self.indptr[i + 1] - 1
            other_row_start = other.indptr[i] - 1
            other_row_end = other.indptr[i + 1] - 1

            self_values = {self.ind[j]: self.data[j] for j in range(self_row_start, self_row_end)}
            other_values = {other.ind[j]: other.data[j] for j in range(other_row_start, other_row_end)}

            for j in range(self.m):
                result_matrix[i][j] = self_values.get(j, 0) + other_values.get(j, 0)

        return Matrix(self.n, self.m, result_matrix)

    def scalar_multiply(self, scalar):
        result_matrix = [[0] * self.m for _ in range(self.n)]

        for i in range(self.n):
            start = self.indptr[i] - 1
            end = self.indptr[i + 1] - 1

            for j in range(start, end):
                result_matrix[i][self.ind[j]] = self.data[j] * scalar

        return Matrix(self.n, self.m, result_matrix)

    def multiply(self, other):
        if self.m != other.n:
            raise ValueError('Количество столбцов первой матрицы не совпадает с количеством строк второй')

        result_matrix = [[0] * other.m for _ in range(self.n)]

        other_values = {}
        for j in range(other.n):
            start = other.indptr[j] - 1
            end = other.indptr[j + 1] - 1
            for k in range(start, end):
                if other.ind[k] not in other_values:
                    other_values[other.ind[k]] = {}
                other_values[other.ind[k]][j] = other.data[k]

        for i in range(self.n):
            self_row_start = self.indptr[i] - 1
            self_row_end = self.indptr[i + 1] - 1
            self_values = {self.ind[j]: self.data[j] for j in range(self_row_start, self_row_end)}

            for k in range(other.m):
                sum_product = 0
                for j in range(self.m):
                    if k in other_values:
                        sum_product += self_values.get(j, 0) * other_values[k].get(j, 0)
                result_matrix[i][k] = sum_product

        return Matrix(self.n, other.m, result_matrix)

    def determinant(self):
        if self.n != self.m:
            raise ValueError("Определитель можно вычислить только для квадратных матриц.")
        if self.n == 2:
            return (self.get_element(1, 1) * self.get_element(2, 2) -
                    self.get_element(1, 2) * self.get_element(2, 1))
        det_value = 0

        def get_minor(row_to_remove, col_to_remove):
            minor_data = []

            for i in range(self.n):
                if i == row_to_remove:
                    continue

                current_row = []
                for j in range(self.m):
                    if j == col_to_remove:
                        continue

                    current_row.append(self.get_element(i + 1, j + 1))

                minor_data.append(current_row)

            return Matrix(len(minor_data), len(minor_data[0]), minor_data)

        for col in range(self.m):
            minor_matrix = get_minor(0, col)
            sign = (-1) ** (col % 2)
            det_value += sign * self.get_element(1, col + 1) * minor_matrix.determinant()

        return det_value
```

### **Метод Гаусса для решения СЛАУ**

### **Основные шаги алгоритма:**

1.   **Проверка входных данных**

*   Матрица А должна быть квадратной
*   Проверка соответствия размеров A и b


2.   **Составление расширенной матрицы**

*   Создание матрицы $[A|b]$ размером $n \cdot (n+1)$

3. **Прямой ход метода Гаусса**

* Выбор ведущего элемента

* Перестановка строк

* Нормализация строк: $a_{ij}= \frac{a_{ij}}{a_{ii}}, (j \geq i)$

* Приведение матрицы к ступенчатому виду: $a_{kj}=a_{kj}-a_{ki} \cdot a_{ij}, (k \neq i)$

4. **Проверка совместности**

* Если найдена строка $[0,0...0][b], b \neq 0,$ то система несовместна

5. **Обратный ход**

* Для единственного решения: $x_i=b_i$ (последний столбец)

* Для бесконечного множества решений: Выделение базисных и свободных переменных, построение ФСР $x_{free}=1; x_{basis}=- \sum a_{ij}x_{j}$

6. **Возврат результатов**

* Список решений (1 элемент для определенной системы, несколько для неопределенной).








```python
def gauss_solver(A: 'Matrix', b: 'Matrix') -> List['Matrix']:
    """
    Вход:
    A: матрица коэффициентов (n×n)
    b: вектор правых частей (n×1)

    Выход:
    list[Matrix]: список базисных векторов решения системы

    Raises:
    ValueError: если система несовместна
    """
    n = A.n

    # проверка размерностей - матрица А должна быть квадратной и вектор b должен быть размера n*1
    if A.n != A.m:
        raise ValueError("Матрица коэффициентов должна быть квадратной")
    if b.n != n or b.m != 1:
        raise ValueError("Несоответствие размеров матрицы и вектора")

    # создаем расширенную матрицу
    expanded_matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            row.append(A.get_element(i+1, j+1))
        row.append(b.get_element(i+1, 1))
        expanded_matrix.append(row)

    # прямой ход метода Гаусса
    rank = 0
    for col in range(n):
        # поиск ведущего элемента
        pivot_row = None
        for row in range(rank, n):
            if expanded_matrix[row][col] != 0:
                pivot_row = row
                break

        if pivot_row is None:
            continue # пропускаем нулевые столбцы

        # перестановка строк для выбора ведущего элемента
        expanded_matrix[rank], expanded_matrix[pivot_row] = expanded_matrix[pivot_row], expanded_matrix[rank]

        # нормализация строки с ведущим элементом
        pivot_val = expanded_matrix[rank][col]
        for j in range(col, n+1):
            expanded_matrix[rank][j] /= pivot_val

        # исключение переменной из других строк
        for row in range(n):
            if row != rank and expanded_matrix[row][col] != 0:
                factor = expanded_matrix[row][col]
                for j in range(col, n+1):
                    expanded_matrix[row][j] -= factor * expanded_matrix[rank][j]

        rank += 1 # увеличиваем ранг после обработки столбца

    # проверка системы на совместность
    for row in range(rank, n): # если в расшир.матрице есть строка [0,0...0][b], b!=0 - то система несовместна
        if abs(expanded_matrix[row][n]) > 1e-10:
            raise ValueError("Система несовместна")

    # обратный ход - нахождение решений
    solutions = [] # создаем список для базисных решений
    basis_indices = [] # индексы базисных переменных

    # определение базисных переменных (столбцы с ведущими элементами)
    for col in range(n):
        for row in range(n):
            if abs(expanded_matrix[row][col] - 1.0) < 1e-10:
                basis_indices.append(col)
                break

    # 1 случай: единственное решение (если rank = n)
    if rank == n and len(basis_indices) == n:
        solution = []
        for i in range(n):
            solution.append([expanded_matrix[i][n]]) # берем последний столбец расшир. матрицы
        return [Matrix(n, 1, solution)]

    # 2 случай: бесконечное множество решений
    free_vars = [i for i in range(n) if i not in basis_indices] # свободные переменные

    # составление базисных решений
    for free_col in free_vars:
        solution = []
        for i in range(n):
            if i in basis_indices:
                for row in range(n): # находим соответствующую строку для базисной переменной
                    if abs(expanded_matrix[row][i] - 1.0) < 1e-10:
                        value = -expanded_matrix[row][free_col] if free_col != i else 1.0
                        solution.append([value])
                        break
            else:
                solution.append([1.0 if i == free_col else 0.0])
        solutions.append(Matrix(n, 1, solution))

    return solutions
```

#### **Тестирование алгоритма Гаусса:**

```python
def test_gauss_solver():
    print("Тест 1: Система с единственным решением")
    A1 = Matrix(3, 3, [
        [2, 1, -1],
        [-3, -1, 2],
        [-2, 1, 2]
    ])
    b1 = Matrix(3, 1, [[8], [-11], [-3]])
    solutions1 = gauss_solver(A1, b1)
    print("Решение:", [[solutions1[0].get_element(i+1, 1) for i in range(3)]])
    # Ожидаемое решение: [2, 3, -1]



    print("\nТест 2: Несовместная система")
    A3 = Matrix(2, 2, [
        [1, 1],
        [1, 1]
    ])
    b3 = Matrix(2, 1, [[1], [2]])
    try:
        gauss_solver(A3, b3)
        print("Ошибка: тест должен вызывать исключение")
    except ValueError as e:
        print(f"Успешно поймано исключение: {e}")

    print("\nТест 3: Диагональная матрица")
    A4 = Matrix(2, 2, [
        [2, 0],
        [0, 3]
    ])
    b4 = Matrix(2, 1, [[4], [9]])
    solutions4 = gauss_solver(A4, b4)
    print("Решение:", [[solutions4[0].get_element(i+1, 1) for i in range(2)]])
    # Ожидаемое решение: [2, 3]

    print("\nТест 4: Матрица с нулевой строкой")
    A5 = Matrix(3, 3, [
        [1, 2, 3],
        [0, 0, 0],
        [7, 8, 9]
    ])
    b5 = Matrix(3, 1, [[6], [0], [24]])
    solutions5 = gauss_solver(A5, b5)
    print("Базисные решения:")
    for sol in solutions5:
        print([[sol.get_element(i+1, 1) for i in range(3)]])
    # Ожидается 1 базисное решение

if __name__ == "__main__":
    test_gauss_solver()
```

### **Функция центрирования данных**

### **Основные шаги алгоритма:**

1. **Создание матрицы из нулей**

* Создается матрица $\text{centered_data}$ того же размера, что и исходная матрица $\text{X}$, заполненная нулями. В этой матрице будут храниться центрированные данные.

2.  **Вычисление среднего значения по каждому столбцу**

* Суммируются все элементы столбца.

* Сумма делится на количество строк в матрице.

3. **Центрирование данных**

* Для каждого элемента в матрице вычитается среднее значение соответствующего столбца.

4. **Запись центрированных данных в матрицу**

* Центрированные данные записываются в матрицу $\text{centered_data}$, которая возвращается как результат.

```python
def center_data(X: 'Matrix') -> 'Matrix':
    """
    Центрирует матрицу данных (вычитает среднее по столбцам).

    Вход: матрица данных X (n×m)
    Выход: центрированная матрица X_centered (n×m)
    """
    n, m = X.n, X.m

    # вычисляем среднее по каждому столбцу
    column_means = [0.0] * m
    for j in range(m):
        column_sum = 0.0
        for i in range(n):
            column_sum += X.get_element(i+1, j+1)
        column_means[j] = column_sum / n if n > 0 else 0.0

    # создаём и заполняем центрированную матрицу
    centered_data = []
    for i in range(n):
        row = []
        for j in range(m):
            original_val = X.get_element(i+1, j+1)
            centered_val = original_val - column_means[j]
            row.append(centered_val)
        centered_data.append(row)

    return Matrix(n, m, centered_data)
```

#### **Тестирование алгоритма нахождения центрированной матрицы:**

### **Вычисление матрицы ковариаций**

### **Основные шаги алгоритма**

**1. Инициализация матрицы ковариаций**

Создается квадратная матрица cov_data размерности m × m, заполненная нулями.

Эта матрица будет хранить вычисленные ковариации между всеми парами признаков.

**2. Вычисление ковариаций для всех пар признаков**

Для каждой пары признаков (i, j) (где i и j пробегают от 0 до m-1):

- Инициализируется переменная sum_products = 0.0 для накопления суммы произведений отклонений.

Для каждого объекта k (от 0 до n-1):

- Получаем значение i-го признака для k-го объекта: x_ki = X_centered[k][i].

- Получаем значение j-го признака для k-го объекта: x_kj = X_centered[k][j].

- Добавляем произведение x_ki * x_kj к sum_products.

Вычисляем ковариацию как sum_products / (n - 1) (несмещенная оценка).

Записываем полученное значение в cov_data[i][j].

**3. Возврат матрицы ковариаций**

Создается объект Matrix на основе cov_data и возвращается как результат.

```python
def covariance_matrix(X_centered: 'Matrix') -> 'Matrix':
    """
    Вычисляет матрицу ковариаций для центрированной матрицы данных.

    Вход: центрированная матрица X_centered (n×m)
    Выход: матрица ковариаций C (m×m)
    """
    n = X_centered.n  # количество объектов
    m = X_centered.m  # количество признаков

    # Создаем пустую матрицу m×m для ковариаций
    cov_data = [[0.0 for _ in range(m)] for _ in range(m)]

    # Вычисляем ковариации между всеми парами признаков
    for i in range(m):
        for j in range(m):
            # Сумма произведений отклонений i-го и j-го признаков
            sum_products = 0.0
            for k in range(n):
                x_ki = X_centered.get_element(k+1, i+1)
                x_kj = X_centered.get_element(k+1, j+1)
                sum_products += x_ki * x_kj
            # Несмещенная оценка ковариации
            cov_data[i][j] = sum_products / (n - 1)

    # Создаем и возвращаем новую матрицу ковариаций
    return Matrix(m, m, cov_data)
```

#### **Тестирование алгоритма нахождения матрицы ковариаций:**

```python
def test_covariance_matrix():
    print("Запуск тестов covariance_matrix...")

    # Тест 1: Простая матрица 2x2
    X_data = [
        [1, 2],
        [3, 4]
    ]
    # Центрируем вручную
    mean_col1 = (1 + 3) / 2
    mean_col2 = (2 + 4) / 2
    X_centered = [
        [1 - mean_col1, 2 - mean_col2],
        [3 - mean_col1, 4 - mean_col2]
    ]
    X_centered_matrix = Matrix(2, 2, X_centered)

    # Ожидаемая ковариационная матрица (вычислено вручную)
    expected_cov = [
        [2.0, 2.0],
        [2.0, 2.0]
    ]

    cov_matrix = covariance_matrix(X_centered_matrix)

    for i in range(2):
        for j in range(2):
            assert abs(cov_matrix.get_element(i+1, j+1) - expected_cov[i][j]) < 1e-9, \
                f"Тест 1 не пройден: элемент [{i},{j}] отличается"

    print("Тест 1 пройден")

    # Тест 2: Матрица 3x1 (один признак)
    X_data = [
        [1],
        [2],
        [3]
    ]
    mean_col = (1 + 2 + 3) / 3
    X_centered = [
        [1 - mean_col],
        [2 - mean_col],
        [3 - mean_col]
    ]
    X_centered_matrix = Matrix(3, 1, X_centered)

    # Дисперсия (ковариация с собой) = сумма квадратов отклонений / (n-1)
    expected_var = ((1-2)**2 + (2-2)**2 + (3-2)**2) / 2

    cov_matrix = covariance_matrix(X_centered_matrix)

    assert abs(cov_matrix.get_element(1, 1) - expected_var) < 1e-9, \
        "Тест 2 не пройден: дисперсия вычислена неверно"

    print("Тест 2 пройден")

    # Тест 3: Матрица 4x3
    X_data = [
        [1, 5, 0],
        [3, 6, 2],
        [2, 7, 1],
        [4, 8, 3]
    ]
    # Вычисляем средние вручную
    means = [
        (1 + 3 + 2 + 4)/4,
        (5 + 6 + 7 + 8)/4,
        (0 + 2 + 1 + 3)/4
    ]
    X_centered = [
        [1-means[0], 5-means[1], 0-means[2]],
        [3-means[0], 6-means[1], 2-means[2]],
        [2-means[0], 7-means[1], 1-means[2]],
        [4-means[0], 8-means[1], 3-means[2]]
    ]
    X_centered_matrix = Matrix(4, 3, X_centered)

    # Вычисляем ожидаемую ковариационную матрицу вручную
    def cov(x, y):
        return sum((x[i] * y[i] for i in range(len(x)))) / (len(x) - 1)

    # Извлекаем столбцы
    col1 = [row[0] for row in X_centered]
    col2 = [row[1] for row in X_centered]
    col3 = [row[2] for row in X_centered]

    expected_cov = [
        [cov(col1, col1), cov(col1, col2), cov(col1, col3)],
        [cov(col2, col1), cov(col2, col2), cov(col2, col3)],
        [cov(col3, col1), cov(col3, col2), cov(col3, col3)]
    ]

    cov_matrix = covariance_matrix(X_centered_matrix)

    for i in range(3):
        for j in range(3):
            assert abs(cov_matrix.get_element(i+1, j+1) - expected_cov[i][j]) < 1e-9, \
                f"Тест 3 не пройден: элемент [{i},{j}] отличается"

    print("Тест 3 пройден")

    # Тест 4: Проверка симметричности матрицы
    for i in range(cov_matrix.n):
        for j in range(cov_matrix.m):
            assert abs(cov_matrix.get_element(i+1, j+1) - cov_matrix.get_element(j+1, i+1)) < 1e-9, \
                f"Тест 4 не пройден: матрица не симметрична на элементах [{i},{j}] и [{j},{i}]"

    print("Тест 4 пройден")
    print("Все тесты успешно пройдены!")

# Запуск тестов
test_covariance_matrix()
```

### **Нахождение собственных значений матрицы C методом бисекции**

```python
def find_eigenvalues(C: 'Matrix', tol: float = 1e-6) -> List[float]:
    """
    Находит собственные значения матрицы методом бисекции.
    Поддерживает кратные собственные значения.

    Параметры:
        C: квадратная матрица (объект класса Matrix)
        tol: точность вычислений (по умолчанию 1e-6)

    Возвращает:
        Список собственных значений в порядке убывания (с учетом кратности)
    """
    if C.n != C.m:
        raise ValueError("Матрица должна быть квадратной")

    n = C.n
    eigenvalues = []

    # Функция для вычисления характеристического полинома
    def char_poly(lambda_):
        # Создаем матрицу (C - λI)
        mat = []
        for i in range(n):
            row = []
            for j in range(n):
                val = C.get_element(i+1, j+1)
                if i == j:
                    val -= lambda_
                row.append(val)
            mat.append(row)
        return Matrix(n, n, mat).determinant()

    # Определяем границы поиска
    row_sums = [sum(abs(C.get_element(i+1,j+1)) for j in range(n)) for i in range(n)]
    a, b = -max(row_sums), max(row_sums)

    # Функция для поиска всех корней на интервале
    def find_roots_on_interval(f, a, b, tol, max_roots=n):
        roots = []
        step = (b - a) / 1000  # Начальный шаг сканирования

        x = a
        while x < b and len(roots) < max_roots:
            x_next = min(x + step, b)
            f_x = f(x)
            f_x_next = f(x_next)

            # Проверяем смену знака или близость к нулю
            if f_x * f_x_next <= 0 or abs(f_x) < tol:
                # Уточняем корень бисекцией
                left, right = x, x_next
                for _ in range(100):  # Ограничиваем итерации
                    mid = (left + right) / 2
                    f_mid = f(mid)

                    if abs(f_mid) < tol or (right - left) < tol:
                        # Проверяем, не нашли ли мы уже этот корень
                        is_new = True
                        for root in roots:
                            if abs(mid - root) < tol * 10:
                                is_new = False
                                break

                        if is_new:
                            roots.append(mid)
                        break

                    if f(left) * f_mid <= 0:
                        right = mid
                    else:
                        left = mid

                # Уменьшаем шаг для поиска близких корней
                step = max(tol, step / 5)
            else:
                # Увеличиваем шаг, если не нашли корень
                step = min((b - a) / 10, step * 2)

            x = x_next

        return roots

    # Ищем все корни характеристического полинома
    eigenvalues = find_roots_on_interval(char_poly, a, b, tol)

    # Учитываем кратность корней
    final_eigenvalues = []
    for val in eigenvalues:
        # Проверяем кратность корня
        # Вычисляем производную численно
        h = tol * 10
        f_val = char_poly(val)
        f_val_plus_h = char_poly(val + h)
        derivative = (f_val_plus_h - f_val) / h

        # Если производная близка к нулю, это кратный корень
        if abs(derivative) < tol:
            final_eigenvalues.extend([val, val])
        else:
            final_eigenvalues.append(val)

    # Сортируем по убыванию и обрезаем до n значений
    final_eigenvalues.sort(reverse=True)
    return final_eigenvalues[:n]
```

```python
# Диагональная матрица 2x2 с известными собственными значениями 3 и 1
diag_matrix = Matrix(2, 2, [[3, 0],
                            [0, 1]])
eigenvalues = find_eigenvalues(diag_matrix)
print(f"Собственные значения диагональной матрицы: {eigenvalues}")
# Ожидаемый результат: [3.0, 1.0] (с точностью до tol)
```

```python
# Матрица 3x3 с кратным собственным значением 2
multi_matrix = Matrix(3, 3, [[2, 1, 0],
                              [0, 2, 1],
                              [0, 0, 2]])
eigenvalues = find_eigenvalues(multi_matrix, tol=1e-5)
print(f"Собственные значения матрицы с кратными значениями: {eigenvalues}")
# Ожидаемый результат: [2.0, 2.0, 2.0] (все одинаковые)
```

```python
# Симметричная матрица 2x2
sym_matrix = Matrix(2, 2, [[4, 1],
                           [1, 2]])
eigenvalues = find_eigenvalues(sym_matrix, tol=1e-4)
print(f"Собственные значения симметричной матрицы: {eigenvalues}")
# Ожидаемый результат (примерно): [4.4142, 1.5858]
```

**Почему метод бисекции плохо подходит для поиска собственных значений**

1. Проблемы с кратными корнями:

Метод бисекции находит корни по смене знака функции

Для кратных корней характеристический полином и его производная обращаются в ноль

График функции "касается" оси X, не пересекая её → бисекция может пропустить корень

2. Низкая скорость сходимости:

Линейная скорость сходимости (точность увеличивается в ~2 раза за каждую итерацию)

Для достижения точности 1e-6 требуется ~20 итераций на каждый корень

3. Проблемы с разделением близких корней:

Если два собственных значения близки (|λ₁ - λ₂| < tol), метод может определить их как один корень

4. Вычислительная сложность:

Каждая итерация требует вычисления определителя → O(n³) операций

Для матрицы n×n нужно найти n корней → общая сложность O(n⁴)

5. Чувствительность к выбору интервала:

Неправильный выбор начального интервала может привести к пропуску корней

**Почему QR-разложение лучше:**

1. Специализированный алгоритм:

Создан именно для поиска собственных значений

2. Преобразует матрицу к удобной форме (верхняя треугольная)

3. Кубическая сходимость:

QR-алгоритм с сдвигами имеет кубическую скорость сходимости

Требует O(n³) операций для всей матрицы (а не для каждого корня)

4. Автоматическая обработка кратных значений:

Корректно находит как простые, так и кратные собственные значения

Не требует анализа производных характеристического полинома

5. Стабильность:

Численно стабилен при правильной реализации

Менее чувствителен к ошибкам округления



### **Алгоритм нахождения собственных значений методом QR-разложения**

Функция find_eigenvalues(C, tol, max_iter) вычисляет собственные значения матрицы ковариаций C с использованием QR-алгоритма.

**Входные данные:**
C — квадратная матрица ковариаций размерности m × m (в CSR-формате).

tol — допустимая погрешность для проверки сходимости (по умолчанию 1e-6).

max_iter — максимальное число итераций (по умолчанию 1000).

**Выходные данные:**
Список вещественных собственных значений, отсортированных по убыванию.

#### **Шаги алгоритма:**

**1. Проверка входных данных**

Убеждаемся, что матрица C квадратная (C.n == C.m), иначе вызываем ошибку.

**2. Инициализация**

Сохраняем исходную матрицу A = C (работаем с копией).

m = A.n — размерность матрицы.

3. Итерационный процесс QR-разложения
Для каждой итерации (max_iter раз):

**3.1. QR-разложение матрицы A**

Выполняем Q, R = qr_decomposition(A), где:

Q — ортогональная матрица,

R — верхняя треугольная матрица.

**3.2. Обновление матрицы A = R @ Q**

Умножаем R на Q (в CSR-формате).

**3.3. Проверка сходимости**

Если все поддиагональные элементы A (нижний треугольник) меньше tol, считаем, что алгоритм сошелся.

Если сходимость достигнута, выходим из цикла.

**4. Извлечение собственных значений**

Собственные значения берутся с диагонали матрицы A.

Сортируем их по убыванию.

### **Алгоритм QR-разложения (метод Грама-Шмидта)**

Функция qr_decomposition(A) выполняет разложение матрицы A в произведение ортогональной Q и верхнетреугольной R.

**1. Подготовка данных**

Создаем плотные (dense) матрицы Q_dense и R_dense размерности m × m, заполненные нулями.

Преобразуем CSR-матрицу A в список столбцов (A_columns).

**2. Ортогонализация Грама-Шмидта**

1. **Берем столбец**  
   `v = A_columns[j]`

2. **Для всех предыдущих столбцов** `k < j`:
   - Вычисляем проекцию v на Q[:,k]:  
     $$R_{k,j} = \sum_{i=1}^{m} Q_{i,k} \cdot A_{i,j}$$

   - Вычитаем проекцию:  

     $$v = v - R_{k,j} \cdot Q[:,k]$$

3. **Нормализация** (если норма > 1e-10):

   $$R_{j,j} = \|v\|, \quad Q[:,j] = \frac{v}{\|v\|}$$  
   где $$\|v\| = \sqrt{\sum_{i=1}^{m} v_i^2}$$

**3. Возврат результатов**

Преобразуем Q_dense и R_dense обратно в объекты Matrix.





```python
def find_eigenvalues(C: 'Matrix', tol: float = 1e-6, max_iter: int = 1000) -> List[float]:
    """
    Находит собственные значения матрицы методом QR-разложения с использованием CSR формата.

    Вход:
    C: матрица ковариаций (m×m) в CSR формате
    tol: допустимая погрешность
    max_iter: максимальное число итераций

    Выход:
    List[float]: список вещественных собственных значений, отсортированных по убыванию
    """
    if C.n != C.m:
        raise ValueError("Матрица должна быть квадратной")

    m = C.n
    A = C  # Работаем с исходной матрицей в CSR формате

    for _ in range(max_iter):
        # QR-разложение
        Q, R = qr_decomposition(A)

        # Обновление матрицы A = R*Q с использованием multiply
        A = R.multiply(Q)

        # Проверка сходимости (нижний треугольник близок к нулю)
        converged = True
        for i in range(1, m):
            for j in range(i):
                if abs(A.get_element(i+1, j+1)) > tol:
                    converged = False
                    break
            if not converged:
                break

        if converged:
            break

    # Извлечение собственных значений с диагонали
    eigenvalues = [A.get_element(i+1, i+1) for i in range(m)]

    # Сортировка по убыванию
    eigenvalues.sort(reverse=True)

    return eigenvalues

def qr_decomposition(A: 'Matrix') -> Tuple['Matrix', 'Matrix']:
    """
    Выполняет QR-разложение матрицы методом Грама-Шмидта для CSR матриц.

    Вход:
    A: квадратная матрица (m×m) в CSR формате

    Выход:
    Q: ортогональная матрица (m×m) в плотном формате
    R: верхняя треугольная матрица (m×m) в плотном формате
    """
    m = A.n
    Q_dense = [[0.0]*m for _ in range(m)]
    R_dense = [[0.0]*m for _ in range(m)]

    # Временное преобразование в плотный формат для столбцов
    A_columns = []
    for j in range(m):
        column = []
        for i in range(m):
            column.append(A.get_element(i+1, j+1))
        A_columns.append(column)

    # Процесс Грама-Шмидта
    for j in range(m):
        v = A_columns[j].copy()

        for k in range(j):
            # Проекция на предыдущие векторы
            R_dense[k][j] = sum(Q_dense[i][k] * A_columns[j][i] for i in range(m))
            v = [v[i] - R_dense[k][j] * Q_dense[i][k] for i in range(m)]

        # Нормализация вектора
        norm = math.sqrt(sum(x**2 for x in v))
        R_dense[j][j] = norm

        if norm > 1e-10:
            for i in range(m):
                Q_dense[i][j] = v[i] / norm

    # Конвертируем обратно в Matrix (плотный формат)
    Q = Matrix(m, m, Q_dense)
    R = Matrix(m, m, R_dense)

    return Q, R
```

### **Нахождение собственных векторов матрицы C**

### **Основные шаги алгоритма:**
 * **Для каждого собственного значения $\lambda$:**

1. Составление матрицы $(C - \lambda I)$.

2. Нахождение базиса нулевого пространства этой матрицы (то есть все векторы $v,$ для которых $(C - \lambda I)v=0$).

3. Каждый такой ненулевой вектор - собственный вектор для $\lambda$.

* **Поиск базиса нулевого пространства:**

1. Приведение матрицы к ступенчатому виду.

2. Определение свободных переменных (столбцы без ведущих единиц).

3. Для каждой свободной переменной построение базисного вектора решения.



```python
def find_eigenvectors(C: 'Matrix', eigenvalues: List[float]) -> List[Tuple[float, 'Matrix']]:
    """
    Находит собственные векторы матрицы C для заданных собственных значений (вещественные числа).

    Аргументы:
        C: квадратная матрица (объект Matrix)
        eigenvalues: список собственных значений (включая кратные)
        tol: допуск для сравнения с нулём

    Возвращает:
        Список кортежей (собственное значение, собственный вектор)
    """
    tol = 1e-8 # допустимая погрешность

    def subtract_lambda_identity(matrix: 'Matrix', lambda_: float) -> 'Matrix':
        """Создает матрицу (matrix - λ*I)"""
        n = matrix.n
        result = []
        for i in range(n):
            row = []
            for j in range(n):
                if i == j: # вычитаем λ только из диагональных эл-тов
                    row.append(matrix.get_element(i+1, j+1) - lambda_)
                else:
                    row.append(matrix.get_element(i+1, j+1))
            result.append(row)
        return Matrix(n, n, result)

    def stepwise_view(matrix: 'Matrix') -> 'Matrix':
        """Приводит матрицу к ступенчатому виду методом Гаусса (с  допустимой погрешностью)"""
        n, m = matrix.n, matrix.m
        mat = [[matrix.get_element(i+1, j+1) for j in range(m)] for i in range(n)]
        rank = 0

        for col in range(m):
            # Поиск ведущего элемента
            pivot = -1
            for row in range(rank, n):
                if abs(mat[row][col]) > tol:
                    pivot = row
                    break

            if pivot == -1:
                continue # пропускаем нулевой столбец

            # Меняем строки местами
            mat[rank], mat[pivot] = mat[pivot], mat[rank]
            pivot_val = mat[rank][col]
            for c in range(col, m): # нормирование строки
                mat[rank][c] /= pivot_val

            # Обнуляем элементы в этом столбце в других строках
            for row in range(n):
                if row != rank and abs(mat[row][col]) > tol:
                    factor = mat[row][col]
                    for c in range(col, m):
                        mat[row][c] -= factor * mat[rank][c]

            rank += 1

        return Matrix(n, m, mat)

    def null_space(matrix: 'Matrix') -> List['Matrix']:
        """Находит базис нуль-пространства матрицы (с допуском)"""
        reduced = stepwise_view(matrix)
        n, m = reduced.n, reduced.m
        pivot_cols = [] # индексы ведущих столбцов

        # Поиск ведущих столбцов (где есть '1' в ступенчатом виде)
        for row in range(n):
            for col in range(m):
                if abs(reduced.get_element(row+1, col+1) - 1.0) < tol:
                    pivot_cols.append(col)
                    break

        free_cols = [col for col in range(m) if col not in pivot_cols] # Свободные переменные - неведущие столбцы
        basis = []

        for free in free_cols: # Построение базисных векторов
            vector = [0.0] * m
            vector[free] = 1.0 # Свободная переменная = 1
            for row in range(len(pivot_cols)): # Обратная подстановка для ведущих переменных
                pivot_col = pivot_cols[row]
                if pivot_col < free:
                    vector[pivot_col] = -reduced.get_element(row+1, free+1)
            basis.append(Matrix(m, 1, [[x] for x in vector]))

        return basis

    # Основной цикл обработки собственных значений
    eigenvectors = []
    for lambda_ in eigenvalues:
        C_minus_lambda_I = subtract_lambda_identity(C, lambda_)
        basis = null_space(C_minus_lambda_I)

        for vec in basis: # сохраняем результаты
            eigenvectors.append((lambda_, vec))

    return eigenvectors
```

```python
def test_find_eigenvectors():
    tol = 1e-6

    def assert_matrix_close(m1, m2, msg=""):
        assert m1.n == m2.n and m1.m == m2.m, f"{msg}: Размеры не совпадают"
        for i in range(1, m1.n + 1):
            for j in range(1, m1.m + 1):
                assert abs(m1.get_element(i, j) - m2.get_element(i, j)) < tol, \
                    f"{msg}: Элементы различаются в позиции ({i},{j})"

    # Тест 1: Диагональная матрица 3x3 (кратность λ=3 равна 2)
    print("Тест 1: Диагональная матрица с кратными с.з.")
    C1 = Matrix(3, 3, [[3,0,0], [0,3,0], [0,0,5]])
    eigenvalues1 = [3.0, 5.0]
    result1 = find_eigenvectors(C1, eigenvalues1)

    # Проверяем количество векторов и условие C*v = λ*v
    assert len(result1) == 3, f"Ожидалось 3 вектора (2 для λ=3, 1 для λ=5), получено {len(result1)}"
    for lambda_, v in result1:
        Cv = C1.multiply(v)
        lambda_v = v.scalar_multiply(lambda_)
        assert_matrix_close(Cv, lambda_v, f"Ошибка для λ={lambda_}")

    # Тест 2: Верхнетреугольная матрица 4x4 (все с.з. различны)
    print("\nТест 2: Верхнетреугольная матрица с разными с.з.")
    B = Matrix(4, 4, [[1,2,3,4],[0,2,3,4],[0,0,3,4],[0,0,0,4]])
    eigenvalues2 = [1.0, 2.0, 3.0, 4.0]
    result2 = find_eigenvectors(B, eigenvalues2)

    assert len(result2) == 4, "Ожидалось 4 вектора"
    for lambda_, v in result2:
        Bv = B.multiply(v)
        lambda_v = v.scalar_multiply(lambda_)
        assert_matrix_close(Bv, lambda_v, f"Ошибка для λ={lambda_}")

    # Тест 3: Матрица с характеристическим полиномом (λ-1)(λ-2)(λ-3)(λ-4)
    print("\nТест 3: Матрица с явно заданными с.з.")
    C3 = Matrix(4, 4, [
        [0, 0, 0, -24],
        [1, 0, 0, 50],
        [0, 1, 0, -35],
        [0, 0, 1, 10]
    ])
    eigenvalues3 = [1.0, 2.0, 3.0, 4.0]
    result3 = find_eigenvectors(C3, eigenvalues3)

    # Проверяем, что векторы не нулевые и удовлетворяют уравнению
    assert len(result3) == 4, "Ожидалось 4 вектора"
    for lambda_, v in result3:
        assert any(abs(v.get_element(i, 1)) > tol for i in range(1, 5)), "Нулевой вектор"
        C3v = C3.multiply(v)
        lambda_v = v.scalar_multiply(lambda_)
        assert_matrix_close(C3v, lambda_v, f"Ошибка для λ={lambda_}")

    # Тест 4: Еще одна диагональная матрица
    print("\nТест 4: Диагональная матрица с разными с.з.")
    C4 = Matrix(3, 3, [[4.5,0,0], [0,3.5,0], [0,0,1]])
    eigenvalues4 = [4.5, 3.5, 1.0]
    result4 = find_eigenvectors(C4, eigenvalues4)

    # Проверяем ортогональность векторов
    assert len(result4) == 3, "Ожидалось 3 вектора"
    vectors = [v for (_, v) in result4]
    for i in range(3):
        for j in range(i+1, 3):
            dot = sum(vectors[i].get_element(k, 1) * vectors[j].get_element(k, 1) for k in range(1, 4))
            assert abs(dot) < tol, f"Векторы для λ={result4[i][0]} и λ={result4[j][0]} не ортогональны"

    print("\nВсе расширенные тесты пройдены успешно!")

if __name__ == "__main__":
    test_find_eigenvectors()
```

### **Вычисление доли объяснённой дисперсии**

$\gamma = \frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^m \lambda_i}$

```python
def explained_variance_ratio(eigenvalues: List[float], k: int) -> float:
    """
    Вычисляет долю объяснённой дисперсии для первых k компонент.

    Вход:
    eigenvalues: список собственных значений, отсортированных по убыванию
    k: количество учитываемых компонент (1 <= k <= len(eigenvalues))

    Выход:
    Доля объяснённой дисперсии (float от 0 до 1)
    """
    if not eigenvalues:
        raise ValueError("Список собственных значений не может быть пустым")
    if k <= 0 or k > len(eigenvalues):
        raise ValueError("k должно быть в диапазоне [1, len(eigenvalues)]")

    sum_k = sum(eigenvalues[:k])  # Сумма первых k собственных значений
    sum_total = sum(eigenvalues)  # Общая сумма всех собственных значений

    return sum_k / sum_total
```

### **Реализация полного алгоритма РСА**

```python
def pca(X: 'Matrix', k: int) -> Tuple['Matrix', float]:
    """
    Полная реализация алгоритма PCA (Principal Component Analysis)

    Вход:
    X: матрица данных (n×m), где n - количество объектов, m - количество признаков
    k: количество главных компонент для проекции (1 ≤ k ≤ m)

    Выход:
    Tuple[Matrix, float]:
      - проекция данных на главные компоненты (n×k)
      - доля объясненной дисперсии
    """
    # 1. Проверка входных данных
    if k <= 0 or k > X.m:
        raise ValueError("k должно быть в диапазоне [1, количество признаков]")

    # 2. Центрирование данных
    X_centered = center_data(X)

    # 3. Вычисление матрицы ковариаций
    C = covariance_matrix(X_centered)

    # 4. Нахождение собственных значений и векторов
    eigenvalues = find_eigenvalues(C)
    print(eigenvalues)
    eigen_pairs = find_eigenvectors(C, eigenvalues)
    print(eigen_pairs)

    # 5. Проверка наличия собственных значений
    if not eigen_pairs:
        raise ValueError("Не найдено ни одного собственного вектора")

    # 6. Сортировка собственных пар по убыванию абсолютных значений
    eigen_pairs.sort(key=lambda x: -abs(x[0]))

    # 7. Проверка количества доступных компонент
    if len(eigen_pairs) < k:
        available = len(eigen_pairs)
        print(f"Предупреждение: доступно только {available} компонент вместо запрошенных {k}")
        k = available

    # 8. Построение матрицы проекции W
    W_data = []
    for i in range(X.m):  # по строкам исходного пространства
        row = []
        for j in range(k):  # по столбцам нового пространства
            row.append(eigen_pairs[j][1].get_element(i+1, 1))
        W_data.append(row)
    W = Matrix(X.m, k, W_data)

    # 9. Проекция данных
    X_proj_data = []
    for i in range(X.n):
        row = []
        for j in range(k):
            val = sum(X_centered.get_element(i+1, l+1) * W.get_element(l+1, j+1) for l in range(X.m))
            row.append(val)
        X_proj_data.append(row)
    X_proj = Matrix(X.n, k, X_proj_data)

    # 10. Расчет доли объясненной дисперсии
    total_variance = sum(abs(val) for val, _ in eigen_pairs)
    explained_variance = sum(abs(val) for val, _ in eigen_pairs[:k])
    ratio = explained_variance / total_variance if total_variance > 1e-10 else 0.0

    return X_proj, ratio
```

### **Вычисление среднеквадратической ошибки восстановления данных**

```python
def reconstruction_error(X_orig: 'Matrix', X_recon: 'Matrix') -> float:
    """
    Вычисляет среднеквадратическую ошибку (MSE) между исходными и восстановленными данными.

    Вход:
    X_orig: исходная матрица данных (n×m)
    X_recon: восстановленная матрица данных (n×m)

    Выход:
    float: значение MSE
    """
    # Проверка совпадения размеров матриц
    if X_orig.n != X_recon.n or X_orig.m != X_recon.m:
        raise ValueError("Размеры матриц должны совпадать")

    total_error = 0.0
    n, m = X_orig.n, X_orig.m

    # Вычисление суммы квадратов разностей
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            diff = X_orig.get_element(i, j) - X_recon.get_element(i, j)
            total_error += diff * diff

    # Вычисление средней ошибки
    mse = total_error / (n * m)

    return mse
```

## **Автоматический выбор числа главных компонент**

```python
def auto_select_k(eigenvalues: List[float], threshold: float = 0.95) -> int:
    """
    Автоматически выбирает число главных компонент на основе порога объяснённой дисперсии.

    Вход:
    eigenvalues: список собственных значений, отсортированных по убыванию
    threshold: порог объяснённой дисперсии (0 < threshold <= 1)

    Выход:
    int: оптимальное число главных компонент k
    """
    if not eigenvalues:
        raise ValueError("Список собственных значений не может быть пустым")
    if threshold <= 0 or threshold > 1:
        raise ValueError("Порог должен быть в диапазоне (0, 1]")

    # Фильтрация очень маленьких значений и проверка на нулевую дисперсию
    abs_eigenvalues = [abs(val) for val in eigenvalues if abs(val) > 1e-10]

    if not abs_eigenvalues:  # Все собственные значения практически нулевые
        return 0

    total_variance = sum(abs_eigenvalues)

    # Проверка на нулевую дисперсию
    if total_variance < 1e-10:
        return 0

    cumulative_variance = 0.0

    for k, val in enumerate(abs_eigenvalues, 1):
        cumulative_variance += val
        if cumulative_variance / total_variance >= threshold:
            return k

    return len(abs_eigenvalues)  # Если не достигли порога, возвращаем все компоненты


def pca(X: 'Matrix', k: Optional[int] = None, threshold: float = 0.95) -> Tuple['Matrix', float, int]:
    """
    Модифицированная функция PCA с автоматическим выбором числа компонент.
    """
    # 1. Центрирование данных
    X_centered = center_data(X)

    # 2. Вычисление ковариационной матрицы
    C = covariance_matrix(X_centered)

    # 3. Нахождение собственных значений и векторов
    eigenvalues = find_eigenvalues(C)
    eigen_pairs = find_eigenvectors(C, eigenvalues)

    # 4. Сортировка по убыванию абсолютных значений
    eigen_pairs.sort(key=lambda x: -abs(x[0]))
    sorted_eigenvalues = [val for val, _ in eigen_pairs]

    # 5. Автоматический выбор k, если не задан
    selected_k = k if k is not None else auto_select_k(sorted_eigenvalues, threshold)

    # 6. Обработка случая, когда все собственные значения нулевые
    if selected_k == 0:
        # Возвращаем нулевую матрицу проекции
        return Matrix(X.n, 0, []), 0.0, 0

    # 7. Проверка доступного числа компонент
    available_k = len(eigen_pairs)
    if available_k < selected_k:
        print(f"Предупреждение: доступно только {available_k} компонент вместо запрошенных {selected_k}")
        selected_k = available_k

    # 8. Построение матрицы проекции W
    W_data = []
    for i in range(X.m):  # по строкам исходного пространства
        row = []
        for j in range(selected_k):  # по столбцам нового пространства
            row.append(eigen_pairs[j][1].get_element(i+1, 1))
        W_data.append(row)
    W = Matrix(X.m, selected_k, W_data)

    # 9. Проекция данных
    X_proj_data = []
    for i in range(X.n):
        row = []
        for j in range(selected_k):
            val = sum(X_centered.get_element(i+1, l+1) * W.get_element(l+1, j+1) for l in range(X.m))
            row.append(val)
        X_proj_data.append(row)
    X_proj = Matrix(X.n, selected_k, X_proj_data)

    # 10. Расчет доли объяснённой дисперсии
    total_variance = sum(abs(val) for val in sorted_eigenvalues)
    if total_variance < 1e-10:
        return X_proj, 0.0, selected_k

    explained_variance = sum(abs(val) for val in sorted_eigenvalues[:selected_k])
    ratio = explained_variance / total_variance

    return X_proj, ratio, selected_k
```

### **Обработка пропущенных значений в данных**

### **Основные шаги алгоритма:**
1. **Сбор данных по столбцам:**

* Проход по каждому элементу матрицы.

* Замена NaN на None для удобства обработки.

* Сохранение значений в виде вложенных списков (отдельный список для каждого столбца).

2. **Вычисление средних значений стобцов:** Для каждого столбца:

* Отфильтровать None.

* Если в столбце есть значения, отличные от None: Вычислить среднее арифметическое.

* Если в столбце все значения None: заменить их на 0.

3. **замена NaN на средние значения:**

* Создание новой матрицы.

* Для каждого элемента исходной матрицы: Если исходное значение NaN - заменить на среднее соответствующего столбца. Иначе - оставить без изменений.




```python
def handle_missing_values(matrix: Matrix) -> Matrix:
    """
    Заменяет пропущенные значения (NaN) в матрице на среднее значение соответствующего столбца.

    Вход: объект Matrix (n×m) с возможными NaN
    Выход: новый объект Matrix (n×m) без NaN

    """
    # собираем данные матрицы по столбцам, заменяя NaN на None
    columns = [[] for _ in range(matrix.m)]

    for i in range(1, matrix.n + 1):
        for j in range(1, matrix.m + 1):
            value = matrix.get_element(i, j)
            # проверяем на NaN (math.isnan работает только с float)
            if isinstance(value, float) and math.isnan(value):
                columns[j-1].append(None)
            else:
                columns[j-1].append(value)

    # вычисляем средние значения для каждого столбца
    column_means = []
    for col in columns:
        # Фильтруем None (которые были NaN) и вычисляем среднее
        valid_values = [x for x in col if x is not None]
        if valid_values:
            mean = sum(valid_values) / len(valid_values)
        else:
            mean = 0  # Если все значения в столбце NaN, заменяем на 0
        column_means.append(mean)

    # создаем новую матрицу с замененными NaN
    filled_matrix = []
    for i in range(matrix.n):
        row = []
        for j in range(matrix.m):
            value = matrix.get_element(i+1, j+1)
            # Если значение NaN, заменяем на среднее
            if isinstance(value, float) and math.isnan(value):
                row.append(column_means[j])
            else:
                row.append(value)
        filled_matrix.append(row)

    # возвращаем новый объект Matrix
    return Matrix(matrix.n, matrix.m, filled_matrix)
```

#### **Тестирование алгоритма замены пропущенных значений:**

```python
def test_handle_missing_values():

    # Тест 1: Матрица без NaN
    m1 = Matrix(2, 2, [[1, 2], [3, 4]])
    result1 = handle_missing_values(m1)
    assert result1.get_element(1, 1) == 1
    assert result1.get_element(2, 2) == 4

    # Тест 2: Матрица с NaN
    m2 = Matrix(3, 2, [[1, 2], [float('nan'), 4], [5, float('nan')]])
    result2 = handle_missing_values(m2)
    # среднее первого столбца (без NaN): (1 + 5)/2 = 3
    assert result2.get_element(2, 1) == 3
    # среднее второго столбца (без NaN): (2 + 4)/2 = 3
    assert result2.get_element(3, 2) == 3

    # Тест 3: Все значения в столбце NaN
    m3 = Matrix(2, 2, [[float('nan'), 2], [float('nan'), 4]])
    result3 = handle_missing_values(m3)
    # первый столбец - все NaN, заменяем на 0
    assert result3.get_element(1, 1) == 0
    assert result3.get_element(2, 1) == 0

    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_handle_missing_values()
```

### **Исследование влияния шума на РСА**

### **Основные шаги алгоритма:**

1. Вычисление стандартного отклонения всех элементов исходной матрицы данных.

2. Добавление к каждому элементу матрицы случайного шума (нормальное распределение с масштабом noise_level $\cdot$ std).

3. Выполнение PCA для исходной матрицы и получение проекции и доли объясненной дисперсии.

4. Выполнение PCA для "зашумленной" матрицы и получение аналогичных результатов.

5. Сравнение результатов: доли объясненной дисперсии и проекции до и после добавления шума.

```python
def add_noise_and_compare(X: 'Matrix', noise_level: float = 0.1) -> dict:
    """
    Добавляет шум к данным и сравнивает результаты PCA до и после добавления шума.

    Вход:
    X: исходная матрица данных (n×m)
    noise_level: уровень шума (доля от стандартного отклонения каждого признака)

    Выход:
    Словарь с результатами сравнения:
    {
        "original_variance_ratio": float,
        "noisy_variance_ratio": float,
        "noise_level": float,
        "projection_diff": float (средняя разница проекций)
    }
    """
    # 1. Вычисляем стандартное отклонение для каждого признака отдельно
    n, m = X.n, X.m
    stds = []
    for j in range(m):
        column = [X.get_element(i+1, j+1) for i in range(n)]
        mean = sum(column) / n
        std = math.sqrt(sum((x - mean)**2 for x in column) / n)
        stds.append(std)

    # 2. Создаем зашумленную матрицу
    noisy_data = []
    for i in range(n):
        row = []
        for j in range(m):
            orig_value = X.get_element(i+1, j+1)
            noise = random.gauss(0, stds[j] * noise_level)
            row.append(orig_value + noise)
        noisy_data.append(row)
    X_noisy = Matrix(n, m, noisy_data)

    # 3. Выполняем PCA для исходных данных
    k = min(2, m)  # Берем не более 2 компонент для сравнения
    try:
        X_proj, orig_ratio, kk = pca(X, k)
    except Exception as e:
        print(f"Ошибка в PCA для исходных данных: {str(e)}")
        return None

    # 4. Выполняем PCA для зашумленных данных
    try:
        X_noisy_proj, noisy_ratio, kk = pca(X_noisy, k)
    except Exception as e:
        print(f"Ошибка в PCA для зашумленных данных: {str(e)}")
        return None

    # 5. Вычисляем среднюю разницу между проекциями
    projection_diff = 0.0
    if hasattr(X_proj, 'from_csr') and hasattr(X_noisy_proj, 'from_csr'):
        orig_proj = X_proj.from_csr()
        noisy_proj = X_noisy_proj.from_csr()
        diff_sum = 0.0
        count = 0
        for i in range(min(len(orig_proj), len(noisy_proj))):
            for j in range(min(len(orig_proj[i]), len(noisy_proj[i]))):
                diff_sum += abs(orig_proj[i][j] - noisy_proj[i][j])
                count += 1
        if count > 0:
            projection_diff = diff_sum / count

    # 6. Формируем результаты
    result = {
        "original_variance_ratio": orig_ratio,
        "noisy_variance_ratio": noisy_ratio,
        "noise_level": noise_level,
        "projection_diff": projection_diff,
        "feature_stds": stds
    }

    # 7. Выводим информацию
    print("\n=== Сравнение PCA ===")
    print(f"Доля объясненной дисперсии (оригинал): {orig_ratio:.4f}")
    print(f"Доля объясненной дисперсии (с шумом): {noisy_ratio:.4f}")
    print(f"Уровень шума: {noise_level}")
    print(f"Средняя разница проекций: {projection_diff:.4f}")
    print("\nСтандартные отклонения признаков:")
    for j, std in enumerate(stds):
        print(f"Признак {j+1}: {std:.4f}")

    return result
```

### **Применение РСА к реальному датасету**

```python
import seaborn as sns
```

```python
def apply_pca_to_dataset(dataset_name: str, k: int) -> Tuple['Matrix', float]:
    """
    Применяет PCA к датасету из Seaborn и возвращает проекцию данных и качество

    Аргументы:
        dataset_name: название датасета ('iris', 'tips', 'titanic' и т.д.)
        k: количество главных компонент

    Возвращает:
        Кортеж (проекция данных как Matrix, качество модели после PCA)
    """
    # 1. Загрузка датасета из Seaborn
    try:
        dataset = "iris"
        dataset = sns.load_dataset(dataset_name)
    except ValueError:
        raise ValueError(f"Датасет {dataset_name} не найден в Seaborn")

    # 2. Преобразование данных в Matrix
    data = []
    for record in dataset.to_dict('records'):
        row = []
        for val in record.values():
            if isinstance(val, (int, float)):
                row.append(float(val))
            elif isinstance(val, str):
                # Простое числовое кодирование строк
                row.append(float(hash(val) % 100))
            else:
                row.append(0.0)
        data.append(row)

    X = Matrix(len(data), len(data[0]), data)

    # 3. Обработка пропущенных значений
    X_clean = handle_missing_values(X)

    # 4. Применение PCA
    X_proj, variance_ratio, kk = pca(X_clean, k)

    # 5. Вычисление качества (упрощенный silhouette score)
    def simple_quality_metric(X: Matrix) -> float:
        n = X.n
        if n <= 1:
            return 0.0

        total_dist = 0.0
        count = 0
        for i in range(1, n+1):
            for j in range(i+1, n+1):
                dist = 0.0
                for dim in range(1, X.m+1):
                    diff = X.get_element(i, dim) - X.get_element(j, dim)
                    dist += diff * diff
                total_dist += math.sqrt(dist)
                count += 1

        return 1.0 / (1.0 + total_dist / count) if count > 0 else 0.0

    # Качество до PCA (по первым k признакам)
    X_before = Matrix(X_clean.n, k, [[X_clean.get_element(i, j+1) for j in range(k)] for i in range(1, X_clean.n+1)])
    quality_before = simple_quality_metric(X_before)

    # Качество после PCA
    quality_after = simple_quality_metric(X_proj)

    print("\nРезультаты PCA:")
    print(f"Доля объясненной дисперсии: {variance_ratio:.4f}")
    print(f"Качество до PCA: {quality_before:.4f}")
    print(f"Качество после PCA: {quality_after:.4f}")

    return X_proj, quality_after
```

### **Выводы:**

* **Доля объясненной дисперсии (0.9953):** Значение близко к 1, значит PCA эффективно сохранил информацию исходных данных, используя всего 2 компоненты. Исходные данные, вероятно, имеют сильную корреляцию между признаками.

* **Качество до PCA (0.4703) vs после PCA (0.0002):** Резкое падение качества (с 0.47 до 0.0002) указывает на проблему в метрике simple_quality_metric: Метрика не адаптирована для оценки PCA, так как использует попарные расстояния между точками, которые могут искажаться после уменьшения размерности. Также PCA не гарантирует сохранение локальных расстояний - он максимизирует глобальную дисперсию.
